```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
import numpy as np
url = "https://raw.githubusercontent.com/TungTh/tungth.github.io/master/data/vn-provinces.json"
try:
    map_data = gpd.read_file(url)
    col_name = 'name' if 'name' in map_data.columns else map_data.columns[1]
    print("Đã tải thành công bản đồ Việt Nam!")
except Exception as e:
    print("Mạng lỗi, không thể tải file GeoJSON. Đang tự động tạo bản đồ giả lập...")
    polygons = []
    names = []
    for i in range(4):
        for j in range(4):
            polygons.append(Polygon([(i, j), (i+1, j), (i+1, j+1), (i, j+1)]))
            names.append(f"Khu vực {i*4 + j + 1}")
    map_data = gpd.GeoDataFrame({'name': names, 'geometry': polygons})
    col_name = 'name'
tinh_list = map_data[col_name].dropna().unique()
df_business = pd.DataFrame({
    col_name: tinh_list,
    'Doanh_Thu': np.random.randint(100, 1000, size=len(tinh_list))
})
merged = map_data.merge(df_business, on=col_name, how="left")
fig, ax = plt.subplots(1, 1, figsize=(10, 12))
merged.plot(
    column='Doanh_Thu',
    ax=ax,
    legend=True,
    cmap='YlGnBu',
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={'label': "Doanh thu (Tỷ đồng)", 'shrink': 0.5}
)

plt.title('23.4: BẢN ĐỒ CHOROPLETH PHÂN BỔ DOANH THU', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import folium
import pandas as pd
import numpy as np
from geopy.distance import geodesic

center_lat, center_lon = 12.2451, 109.1943
center_location = (center_lat, center_lon)
np.random.seed(42)
num_customers = 150
latitudes = center_lat + np.random.uniform(-0.1, 0.1, num_customers)
longitudes = center_lon + np.random.uniform(-0.1, 0.1, num_customers)
customers = pd.DataFrame({
    'CustomerID': range(1, num_customers + 1),
    'Lat': latitudes,
    'Lon': longitudes
})
def calculate_zone(row):
    dist_km = geodesic(center_location, (row['Lat'], row['Lon'])).kilometers
    if dist_km <= 3:
        return 'Vùng 1 (0-3km)', dist_km, 'green'
    elif dist_km <= 5:
        return 'Vùng 2 (3-5km)', dist_km, 'orange'
    elif dist_km <= 10:
        return 'Vùng 3 (5-10km)', dist_km, 'blue'
    else:
        return 'Ngoài vùng (>10km)', dist_km, 'red'

customers[['Zone', 'Distance_km', 'Color']] = customers.apply(
    lambda row: pd.Series(calculate_zone(row)), axis=1
)
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)
folium.Marker(
    location=[center_lat, center_lon],
    popup="<b>KHO HÀNG KHÁNH HÒA</b>",
    icon=folium.Icon(color="darkred", icon="home", prefix='fa')
).add_to(m)
radii_config = {
    3000: {'color': 'green', 'name': '3 km'},
    5000: {'color': 'orange', 'name': '5 km'},
    10000: {'color': 'blue', 'name': '10 km'}
}
for radius_m, config in radii_config.items():
    folium.Circle(
        location=[center_lat, center_lon],
        radius=radius_m,
        color=config['color'],
        weight=2,
        fill=True,
        fill_opacity=0.1,
        popup=f"Bán kính phục vụ: {config['name']}"
    ).add_to(m)
for _, row in customers.iterrows():
    folium.CircleMarker(
        location=[row['Lat'], row['Lon']],
        radius=4,
        color=row['Color'],
        fill=True,
        fill_opacity=0.7,
        popup=f"KH {row['CustomerID']}<br>Cách kho: {row['Distance_km']:.2f} km"
    ).add_to(m)
m
```

```python
import osmnx as ox
import networkx as nx
import random
import matplotlib.pyplot as plt

place_name = "Dien Khanh District, Khanh Hoa, Vietnam"
G = ox.graph_from_place(place_name, network_type='drive')
nodes = list(G.nodes())
start_node = random.choice(nodes)
end_node = random.choice(nodes)
A = nx.shortest_path(G, start_node, end_node, weight='length')
B = nx.astar_path(G, start_node, end_node, weight='length')

print(f"Đường đi A: {A}")
print(f"Đường đi B: {B}")
fig, ax = ox.plot_graph_routes(
    G,
    routes=[A, B],
    route_colors=['red', 'blue'],
    route_linewidths=[6, 2],
    node_size=0
)
print(f"Độ dài A: {nx.path_weight(G, A, weight='length'):.2f} mét")
print(f"Độ dài B: {nx.path_weight(G, B, weight='length'):.2f} mét")

print("\nPhân tích (Bối cảnh Diên Khánh):")
print("- Dijkstra: Quét mọi hướng để tìm đường ngắn nhất, hiệu quả cho các mạng lưới đường phức tạp quanh Thành cổ.")
print("- A*: Định hướng tìm kiếm về phía đích, giúp tiết kiệm tài nguyên tính toán khi xử lý các tuyến đường dài nối các xã.")
```

```python
import folium
import osmnx as ox
import networkx as nx
center = [10.7769, 106.7009]
customer_loc = [10.7700, 106.7100]
car_loc = [10.7800, 106.6900]
print("Đang tải dữ liệu bản đồ...")
G = ox.graph_from_point(tuple(center), dist=3000, network_type='drive')
orig_node = ox.distance.nearest_nodes(G, X=car_loc[1], Y=car_loc[0])
dest_node = ox.distance.nearest_nodes(G, X=customer_loc[1], Y=customer_loc[0])
route = nx.shortest_path(G, orig_node, dest_node, weight='length')
route_length = nx.shortest_path_length(G, orig_node, dest_node, weight='length')
print(f"Khoảng cách đường đi thực tế: {route_length:.2f} mét")
route_coords = []
for node in route:
    lat = G.nodes[node]['y']
    lon = G.nodes[node]['x']
    route_coords.append([lat, lon])
m = folium.Map(location=center, zoom_start=14)
folium.Marker(customer_loc, icon=folium.Icon(color='blue', icon='user'), popup="Khách hàng").add_to(m)
folium.Marker(car_loc, icon=folium.Icon(color='red', icon='car'), popup="Xe công nghệ").add_to(m)
folium.PolyLine(route_coords, color="red", weight=4, opacity=0.8, popup=f"{route_length:.2f} mét").add_to(m)
m
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import folium
np.random.seed(42)
n_customers = 100
data = {
    'lat': np.random.uniform(10.75, 10.85, n_customers),
    'lon': np.random.uniform(106.6, 106.75, n_customers)
}
df = pd.DataFrame(data)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(df[['lat', 'lon']])
centroids = kmeans.cluster_centers_
m = folium.Map(location=[10.8, 106.67], zoom_start=12)
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=3,
        color='blue',
        fill=True,
        fill_opacity=0.6
    ).add_to(m)
for i, coord in enumerate(centroids):
    folium.Marker(
        location=[coord[0], coord[1]],
        popup=f"Trạm/Kho hàng {i+1} (Tối ưu)",
        icon=folium.Icon(color='red', icon='star')
    ).add_to(m)
print("Đã tạo bản đồ trực quan hóa vị trí tối ưu!")
m
```

```python
import folium
from folium import plugins
data_roads = [
    {"coords": [[10.762, 106.682], [10.772, 106.692]], "name": "Đường A", "density": 0.9, "speed": 10}, # Nguy cơ cao
    {"coords": [[10.772, 106.692], [10.782, 106.702]], "name": "Đường B", "density": 0.3, "speed": 45}, # Nguy cơ thấp
    {"coords": [[10.762, 106.682], [10.752, 106.672]], "name": "Đường C (Dự phòng)", "density": 0.2, "speed": 50}
]
def calculate_risk(density, speed):
    """Hàm logic mờ đơn giản hóa để xác định nguy cơ"""
    if density > 0.7 and speed < 20:
        return "High", "red"
    elif density > 0.4 or speed < 35:
        return "Medium", "orange"
    else:
        return "Low", "green"
m = folium.Map(location=[10.7626, 106.6844], zoom_start=14)
for road in data_roads:
    risk_level, color = calculate_risk(road['density'], road['speed'])
    folium.PolyLine(
        locations=road['coords'],
        color=color,
        weight=8,
        opacity=0.8,
        tooltip=f"{road['name']} - Nguy cơ: {risk_level}"
    ).add_to(m)
    if risk_level == "High":
        alt_road = data_roads[2]
        folium.PolyLine(
            locations=alt_road['coords'],
            color='blue',
            weight=5,
            dash_array='10',
            tooltip="Tuyến đường thay thế đề xuất"
        ).add_to(m)

        folium.Marker(
            location=alt_road['coords'][0],
            icon=folium.Icon(color='blue', icon='info-sign'),
            popup="Lối rẽ gợi ý để tránh tắc nghẽn"
        ).add_to(m)
print("Bản đồ phân tích nguy cơ đã được tạo!")
m
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import folium
from folium.plugins import HeatMap
data = pd.DataFrame({
    'zone_id': np.repeat(range(5), 10),
    'hour': np.tile(range(10), 5),
    'demand': np.random.poisson(lam=20, size=50)
})
X = data[['zone_id', 'hour']]
y = data['demand']
model = RandomForestRegressor(n_estimators=100).fit(X, y)
zones_coords = {
    0: [10.77, 106.69], 1: [10.78, 106.70], 2: [10.76, 106.68],
    3: [10.75, 106.67], 4: [10.79, 106.71]
}
predict_data = pd.DataFrame({'zone_id': range(5), 'hour': 11})
predictions = model.predict(predict_data)
m = folium.Map(location=[10.77, 106.69], zoom_start=13)
heat_data = [[zones_coords[i][0], zones_coords[i][1], predictions[i]] for i in range(5)]
HeatMap(heat_data, radius=25).add_to(m)
m
```

```python
import folium
import numpy as np
depot = [10.775, 106.700]
customers = [[10.782, 106.695], [10.770, 106.685], [10.760, 106.710], [10.790, 106.720]]
def euclidean_distance(p1, p2):
    return np.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
def solve_vrp(start, points):
    route = [start]
    unvisited = points.copy()
    while unvisited:
        nearest = min(unvisited, key=lambda p: euclidean_distance(route[-1], p))
        route.append(nearest)
        unvisited.remove(nearest)
    route.append(start)
    return route
optimized_route = solve_vrp(depot, customers)
m = folium.Map(location=depot, zoom_start=14)
folium.Marker(depot, icon=folium.Icon(color='red', icon='home')).add_to(m)
for c in customers:
    folium.Marker(c, icon=folium.Icon(color='blue')).add_to(m)
folium.PolyLine(optimized_route, color="green", weight=5, opacity=0.7, tooltip="Tuyến tối ưu").add_to(m)
m
```
